# Pandas
Empezamos con la librería "pandas". Tal y como hemos hablado, pandas es una librería de estructura y análisis de datos. La principal estructura de datos que utiliza pandas es el Dataframe.
## Crear un DataFrame
Existen varias formas de crear un dataframe. Podemos crearlo desde un diccionario:

```python
dic = {'a': [0,1,2,3,4], 'b': [4,5,6,1,1]}
df = pd.DataFrame(data=dic)
print(df)
```

O un array:

```python
array = [0,3,4,52,4,45,7,8]
df2 = pd.DataFrame(data=array)
print (df2)
```

También se puede crear desde un fichero ya existente que contenga la información que necesites procesar. Por ejemplo un csv, un JSON o incluso una tabla en SQL.

## Enredando con las columnas
### Eliminar columnas

```python
df.drop('Agriculture', axis=1, inplace=True)
df.drop('Industry', axis=1, inplace=True)
df.drop('Service', axis=1, inplace=True)
df.info()
```

### Crear y cambiar columnas
Si, por ejemplo, no nos gusta el nombre que tiene la columna del área del país y queremos que el nombre sea "Area", sin lo que hay entre paréntesis, deberíamos hacer lo siguiente:

```python
df['Area']= df['Area (sq. mi.)']
df.drop('Area (sq. mi.)', axis=1, inplace=True)
df.info()
```

## Consultas en nuestros datos

```python
df['Area'].max()
```

### ¿Qué region tiene más población?

```python
p = df.groupby('Region')
p = p.sum()
p = p.sort_values(['Population'], ascending=False)
p
```

### ¿Cuál es el país con menor PIB?

```python
pib = df.sort_values(['GDP ($ per capita)'])
pib.head(3)
```

## ¿Y qué pasa con las series?
Las series son la estructura de datos de pandas de una sola dimensión (además del índice). Se utilizan, entre otras cosas, para las series temporales, en la que la única información que tenemos será un dato por cada día/hora/mes... lo que sea. DataFrame y Series comparten muchas funciones, aunque hay algunas propias de cada estructura. Vamos a echar un vistazo rápido a las series. Utilizamos un dataset de nacimientos diarios en una ciudad de EEUU en el año 1959. 

```python
series= pd.Series.from_csv("daily-total-female-births.csv", header=0)
series.head(10)
```

Se puede consultar por el índice exactamente igual que si es un indice numérico. En este caso, el índice es de tipo DateTime, pero puede ser consultado como una cadena.

```python
series['1959-07-01']
```

## Remuestreo 

```python
series = series.resample('W').sum()
series.head(10)
```

```python
series=series.resample('A').mean()
series
```

# Matplotlib
Esta es una librería de visualización de datos, podemos realizar gráficas de cualquier tipo y personalizarla como queramos. Su página web tiene muchos ejemplos y una documentación muy completa. Puedes consultarla aquí: https://matplotlib.org/
## Line Plot

```python

fig, ax = plt.subplots()
ax.plot(series)
ax.set(xlabel='Fecha', ylabel='Nacimientos', title='Mi lineplot')
ax.grid()
plt.show()
```

## Estilos de la gráfica

```python
print(plt.style.available)
```

```python
plt.style.use('grayscale')
fig, ax = plt.subplots()
ax.plot(series)
ax.set(xlabel='Fecha', ylabel='Nacimientos', title='Mi lineplot')
ax.grid()
plt.show()
```

## Cambiar las etiquetas

```python
fig, ax = plt.subplots()
ax.plot(series)
ax.set(xlabel='Fecha', ylabel='Nacimientos', title='Mi lineplot')
labels= ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.grid()
plt.show()
```
